```python
#!/usr/bin/python
# -*- coding:utf-8 -*-
# author: Jola
# CreateTime: 2018/12/07 20:10
# software-version: python 3.6


import requests
from scrapy.selector import Selector
from fake_useragent import UserAgent


def crawler_ips():
    """爬取代理池IP

    最好采用代理池爬，爬了1397页之后被封了IP
    有 socket4/5 的IP，暂时没保存，需要时改代码
    过滤了5秒以上的代理IP

    :return:None
    """
    all_proxies = []

    headers = {
#         'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:52.0) Gecko/20100101 Firefox/52.0'
        'User-Agent': UserAgent().random
    }
    url = 'http://www.xicidaili.com/nn/{0}'

    for i in range(1, 10):

        print('=========No ', i)
        req = requests.get(url.format(i), headers=headers)

        if req.status_code != 200:
            continue
        selector = Selector(text=req.text)
        all_ips = selector.css('#ip_list tr')
        current_proxies = []
        ip_lists = []
        for each_ip in all_ips[1:]:
            speed_str = each_ip.css('.bar::attr(title)').extract()[0]
            speed = float(speed_str.split('秒')[0])
            if speed > 3:
                continue
            all_texts = each_ip.css('td::text').extract()
            ip = all_texts[0]
            port = all_texts[1]
            proxy_type = all_texts[5].replace('\n', '').replace(' ', '') or all_texts[4]
            if proxy_type != 'HTTPS'and proxy_type != 'HTTP':
                continue
            ip_lists.append((ip, port, speed, proxy_type))

        for each in ip_lists:
            try:
                result = requests.get('https://www.baidu.com', proxies={each[3]: str(each[0]+str(each[1]))}, timeout=2)
            except Exception as e:
                continue
            if result.status_code < 200 or result.status_code > 300:
                continue
            current_proxies.append({each[3]: str(each[0]+str(each[1]))})

        all_proxies.extend(current_proxies)

    print(all_proxies)

crawler_ips()
```
