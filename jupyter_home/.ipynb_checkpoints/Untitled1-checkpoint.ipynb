```python
import os
#import numpy as np

os.chdir("D:/OneDrive/study/wustl/intro to python/midterm_project")

Knapsack_Instances_file = "Knapsack_Instances.csv"
f = open(Knapsack_Instances_file, "r")
lines = f.readlines()

#print(lines)

weight_list = []
value_list = []

for i in range(len(lines)):
    if i % 3 == 0:
        if lines[i].strip('\n').split(',') != ['']:
            weight_list.extend(lines[i].strip('\n').split(','))
    else:
        if lines[i].strip('\n').split(',') != ['']:
            value_list.extend(lines[i].strip('\n').split(','))
            
#print(len(weight_list) == len(value_list))
item_list = []
index = 0
for x in weight_list:
    while index != len(value_list):
        item_list.append([int(x),float(value_list[index])])
        index += 1
        break
```

```python
capacity = 100

def knapsack(item_list, capacity):
    capacity_list = [[0 for i in range(capacity + 1)] for j in range(len(item_list))]
    cur_item = 0
    weight = 0
    value = 1

    while cur_item != len(item_list):
        cur_capacity = 0
        while cur_capacity != len(capacity_list[0]):
            if item_list[cur_item][weight] <= cur_capacity:
                capacity_list[cur_item][cur_capacity] = max(capacity_list[cur_item - 1][cur_capacity],
                                 item_list[cur_item][value] + capacity_list[cur_item - 1][cur_capacity - item_list[cur_item][weight]])
            else:
                capacity_list[cur_item][cur_capacity] = capacity_list[cur_item - 1][cur_capacity]
            cur_capacity += 1
        cur_item += 1
    return [capacity_list[-1][-1],capacity_list]
```

```python
def choice_list(item_list, capacity_list):
    weight = 0
    
    added_list = [0 for i in range(len(item_list))]
    max_item = len(item_list) - 1
    max_capacity = len(capacity_list[0]) - 1

    while max_item > -1 and max_capacity > 0:
        if max_item - 1 < 0:
            added_list[max_item] = 1
        if capacity_list[max_item][max_capacity] > capacity_list[max_item - 1][max_capacity]:
            added_list[max_item] = 1
            max_capacity = max_capacity - item_list[max_item][weight]
        max_item = max_item - 1   
    return added_list   

decision_list = choice_list(item_list,knapsack(item_list,capacity)[1])
```

```python
new_decision_list = []
temp1 = []
for index in range(len(decision_list)):
    
    if (index + 1) % 50 == 0:
        temp1.append(decision_list[index])
        new_decision_list.append(temp1)
        temp1 = []
    else:
        temp1.append(decision_list[index])
    
decision_str = []

for i in range(len(new_decision_list)):
    decision_str.append( ', '.join([str(j) for j in new_decision_list[i]]) + '\n')



new_weight_list = []
temp2 = []
for index in range(len(weight_list)):
    
    if (index + 1) % 50 == 0:
        temp2.append(weight_list[index])
        new_weight_list.append(temp2)
        temp2 = []
    else:
        temp2.append(weight_list[index])
    
weight_str = []
for i in range(len(new_weight_list)):
    weight_str.append( ', '.join([str(j) for j in new_weight_list[i]]) + '\n')



new_value_list = []
temp3 = []
for index in range(len(value_list)):
    
    if (index + 1) % 50 == 0:
        temp3.append(value_list[index])
        new_value_list.append(temp3)
        temp3 = []
    else:
        temp3.append(value_list[index])
    
value_str = []
for i in range(len(new_value_list)):
    value_str.append( ', '.join([str(j) for j in new_value_list[i]]) + '\n')



final_excel = []
for index in range(len(weight_str)):
    final_excel.append(weight_str[index])
    final_excel.append(value_str[index])
    final_excel.append(decision_str[index])
        


test_knapsack_solution = "test_knapsack_solution.csv"
f = open(test_knapsack_solution, "w")

f.writelines(final_excel)
f.close()
```
