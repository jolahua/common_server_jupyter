```python
#!/usr/bin/python
# -*- coding:utf-8 -*-
# author: Jola
# CreateTime: 2018/10/9 10:34
# software: PyCharm


import pandas as pd
import numpy as np
import time
import xlwt
import os


def generate_knap_size():
    """
    随机生成背包大小
    :return:
    """
    f = open("Knap_Sizes.csv", "w")

    knap_sizes = np.random.uniform(100, 150, 500)
    for size in knap_sizes:
        f.write("%0.2f\n" % size)

    f.close()
    return knap_sizes


def read_csv(instance_file_path, size_path=None):
    """
    把背包重量和价值 以及 所有背包的大小读取出来
    :param instance_file_path: 价值重量文件
    :param size_path: 背包大小文件
    :return:
    """
    instance_list = pd.read_csv(instance_file_path, encoding='utf-8', header=None)
    # print(instance_list.shape)
    v_list = []
    w_list = []

    for n in range(instance_list.shape[0]):
        if (n + 1) % 3 == 1:  # 判断weight 还是 value
            w_list.append(list(instance_list.iloc(0)[n]))
        elif (n + 1) % 3 == 2:
            v_list.append(list(instance_list.iloc(0)[n]))
    if size_path:
        size_list = pd.read_csv(size_path, encoding='utf-8', header=None)
        c_list = list(size_list[0])
        return v_list, w_list, c_list

    return v_list, w_list


def bag(n, c, w, v):
    """
    背包问题  动态规划实现

    :param n: 背包可容纳物品数
    :param c: 背包允许最大重量
    :param w: 待选取的重量列表
    :param v: 待选取的价值列表
    :return: 返回包含所有可能的数组
    """
    res = [[-1 for _ in range(c + 1)] for _ in range(n + 1)]    # 生成动态数组
    for j in range(c + 1):  # 这一步可以不要  初始化为0
        res[0][j] = 0
    for i in range(1, n + 1):
        for j in range(1, c + 1):
            res[i][j] = res[i - 1][j]   # 当前背包的重量
            if j >= w[i - 1] and res[i][j] < res[i - 1][j - w[i - 1]] + v[i - 1]:   # 判断是否价值最优
                res[i][j] = res[i - 1][j - w[i - 1]] + v[i - 1]
    return res


def show(n, c, w, matrix):
    """
    获取当前数组的索引 需要写入文件

    :param n: 背包可容纳物品数
    :param c: 背包允许最大重量
    :param w: 待选取的重量列表
    :param matrix: 价值数组
    :return: 返回一个 0 1 数组 可用作csv生成
    """
    # print('最大价值为:', matrix[n][c])
    chooice_list = []
    x = [False for i in range(n)]
    j = c
    for i in range(n, 0, -1):
        if matrix[i][j] > matrix[i - 1][j]:
            x[i - 1] = True
            j -= w[i - 1]
    # print('背包中所装物品为:')
    for i in range(n):
        if x[i]:
            # print('第', i + 1, '个,', end='')
            chooice_list.append(1)
        else:
            chooice_list.append(0)

    # print(chooice_list)
    return chooice_list


def to_csv(v_list, w_list, chooice_list, file_name='52_midterm.csv'):
    """
    生成结果csv
    :param w_list: 重量
    :param v_list: 价值
    :param chooice_list: 01 列表
    :param file_name: 文件名称
    :return:
    """
    csv = xlwt.Workbook(encoding='utf-8')
    sheet = csv.add_sheet('sheet1')
    for i in range(len(w_list) * 3):
        for j in range(len(w_list[0])):
            if (i + 1) % 3 == 1:
                sheet.write(i, j, w_list[int(i / 3)][j])
            elif (i + 1) % 3 == 2:
                sheet.write(i, j, v_list[int(i / 3)][j])
            elif (i + 1) % 3 == 0:
                sheet.write(i, j, chooice_list[int(i / 3)][j])
    csv.save(file_name)
    print('Programing Finish..\nCsv Generate Finish..')


def main(generate=False):
    """
    主程序入口
    :param generate: 如果需要每次生成随机大小，则传入generate
    :return:
    """
    instance_file_path = 'Knapsack_Instances.csv'
    size_path = 'Knap_Sizes.csv'

    if generate or (not os.path.exists('Knap_Sizes.csv')):
        c_list = generate_knap_size()
        v_list, w_list = read_csv(instance_file_path)
    else:
        v_list, w_list, c_list = read_csv(instance_file_path, size_path)
    n = 50
    chooice_list = []
    for each in range(len(v_list)):    # 循环所有背包问题
        print('=======current_num', each)
        v = [round(e * 100) for e in v_list[each]]
        w = [round(e * 100) for e in w_list[each]]
        c = round(c_list[each] * 100)
        matrix = bag(n, c, w, v)
        chooice_list.append(show(n, c, w, matrix))
    # print(v_list, w_list, chooice_list)
    to_csv(v_list, w_list, chooice_list)


```

```python

# main(True)    # 每次运行都随机生成
main()
```
