````python
def transfer_model(pretrained_model='vgg16'):
    '''
    Create model with pretrained model for transfer learning
    
    Arguments:
    pretrained_model -- string, name of pretrained model
    
    Return:
    model -- object, model created
    '''
    # create model
    cmd = 'model = models.' + pretrained_model + '(pretrained=True)'
    print(cmd)
    exec(cmd, globals())
    
    # Freeze parameters so we don't backprop through them
    for param in model.parameters():
        param.requires_grad = False
        
    # change classifier for models
    if pretrained_model[:3] == 'vgg':
            classifier = nn.Sequential(OrderedDict([
                                                  ('fc1', nn.Linear(25088, 4096)),
                                                  ('relu1', nn.ReLU()),
                                                  ('dropout1', nn.Dropout(0.5)),
                                                  ('fc2', nn.Linear(4096, 768)),
                                                  ('relu2', nn.ReLU()),
                                                  ('dropout2', nn.Dropout(0.5)),
                                                  ('fc3', nn.Linear(768, 102)),
                                                  ('output', nn.LogSoftmax(dim=1))
                                                  ]))    
    model.classifier = classifier
    
    # Add attributes to model
    model.pretrained_model = pretrained_model
    model.epoch = 0
    
    return model
````

````python
# create model with vgg16
model = transfer_model('vgg16')
model = model.to(device)
````

````python
# Define function for validation
def validation(model, validloader, criterion, device='cpu'):
    '''
    Test model on testing data, measure loss and accuracy(correct/total)
    
    Notes:
    This function will turn off dropout
    
    Arguments:
    model -- object, network to be tested
    validloader -- object, loader of validation data
    criterion -- object, loss function
    
    Return:
    loss -- number, average loss of all validation data
    accuracy -- number, average accuracy of all validation data(correct/total)
    '''
    # Initialize variables
    valid_loss = 0
    accuracy = 0
    len_validloader = len(validloader)
    
    # Turn off dropout to evaluate full network
    model.eval()
                
    # Evaluate model on validation data
    with torch.no_grad():
        # Get and reshape data
        for images, labels in validloader:
            images, labels = images.to(device), labels.to(device)
        
            # forward pass
            output = model.forward(images)
        
            # measure loss
            valid_loss += criterion(output, labels).item()
        
            # measure accuracy
            # Model's output is log-softmax, take exponential to get the probabilities
            ps = torch.exp(output)
            # Class with highest probability is our predicted class, compare with true label
            equality = (labels.data == ps.max(1)[1])
            # Accuracy is number of correct predictions divided by all predictions, just take the mean
            accuracy += equality.type_as(torch.FloatTensor()).mean()
    
    return valid_loss/len_validloader, accuracy/len_validloader


# Define function for training
def train(model, trainloader, validloader, criterion, optimizer, epochs=6, print_every=40, device='cpu'):
    '''
    Train model on training data, measure loss and accuracy
    
    Argument:
    model -- object, network to be trained
    trainloader -- object, loader of training data
    validloader -- object, loader of validation data
    criterion -- object, loss function
    optimizer -- object, optimization function
    epochs -- integer, number of training rounds, default=2
    print_every -- integet, number of training steps to display measurements, default=40
    
    Return:
    '''
    # Initialize variables
    train_loss = 0
    steps = 0
    
    # Turn on dropout
    model.train()
    
    # Loop for each epoch
    for e in range(epochs):
        
        # Get and reshape data
        for images, labels in trainloader:
            images, labels = images.to(device), labels.to(device)
            
            # update step count
            steps += 1
            
            # clear the gradients for optimizer
            optimizer.zero_grad()
            
            # forward pass, calculate loss, backword pass, update weights and bias
            output = model.forward(images)
            loss = criterion(output, labels)
            loss.backward()
            optimizer.step()
            
            # measure loss
            train_loss += loss.item()
            
            # Evaluate and display measurements every 'print_every' steps
            if (steps % print_every) == 0:                
                # Evaluate model on validation data
                valid_loss, accuracy = validation(model, validloader, criterion, device=device)
                
                # Display measurements
                print('Epoch:{}/{}  '.format(e+1, epochs),
                      'Train Loss={:.3f}  '.format(train_loss/print_every),
                      'Validation Loss={:.3f}  '.format(valid_loss),
                      'Validation Accuracy={:.2f} %'.format(accuracy*100))
                
                # reset train_loss
                train_loss = 0
                
                # Turn on dropout and grad
                model.train()
        
        model.epoch += 1
````

````python
# Define criterion and optimizer
criterion = nn.NLLLoss()
criterion.definition = 'criterion = nn.NLLLoss()'

# Only train the classifier parameters, feature parameters are frozen
optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)
optimizer.definition = 'optimizer = optim.Adam(model.classifier.parameters(), lr=0.001)'

start = time.time()
train(model, trainloader, testloader, criterion, optimizer, epochs=6, device=device)
print(f"Device = {device}; Time of training: {(time.time() - start):.3f} seconds")
````

## Loading the checkpoint

At this point it's good to write a function that can load a checkpoint and rebuild the model. That way you can come back to this project and keep working on it without having to retrain the network.

````python
# TODO: Write a function that loads a checkpoint and rebuilds the model

def load_checkpoint(file_path, device='cpu'):
    '''
    Load checkpoint from file_path to recover model, criterion, optimizer
    
    Arguments:
    file_path -- string, path of file previously saving the checkpoint
    
    Return:
    None
    '''
    # Load checkpoint from file
    checkpoint = torch.load(file_path)
    
    # Re-create model and recover itw state
    model = transfer_model(checkpoint['pretrained_model'])
    model.epoch = checkpoint['epoch']
    model.class_to_idx = checkpoint['class_to_idx']
    model.load_state_dict(checkpoint['model_state_dict'])
    model = model.to(device)

    # Re-create criterion
    print(checkpoint['criterion_definition'])
    exec(checkpoint['criterion_definition'], globals())
    criterion.definition = checkpoint['criterion_definition']

    # Re-create optimizer and recover its state
    print(checkpoint['optimizer_definition'])
    exec(checkpoint['optimizer_definition'], globals())
    optimizer.load_state_dict = checkpoint['optimizer_state_dict']
    optimizer.definition = checkpoint['optimizer_definition']
````

## Image Preprocessing

You'll want to use `PIL` to load the image ([documentation](https://pillow.readthedocs.io/en/latest/reference/Image.html)). It's best to write a function that preprocesses the image so it can be used as input for the model. This function should process the images in the same manner used for training. 

First, resize the images where the shortest side is 256 pixels, keeping the aspect ratio. This can be done with the [`thumbnail`](http://pillow.readthedocs.io/en/3.1.x/reference/Image.html#PIL.Image.Image.thumbnail) or [`resize`](http://pillow.readthedocs.io/en/3.1.x/reference/Image.html#PIL.Image.Image.thumbnail) methods. Then you'll need to crop out the center 224x224 portion of the image.

Color channels of images are typically encoded as integers 0-255, but the model expected floats 0-1. You'll need to convert the values. It's easiest with a Numpy array, which you can get from a PIL image like so `np_image = np.array(pil_image)`.

As before, the network expects the images to be normalized in a specific way. For the means, it's `[0.485, 0.456, 0.406]` and for the standard deviations `[0.229, 0.224, 0.225]`. You'll want to subtract the means from each color channel, then divide by the standard deviation. 

And finally, PyTorch expects the color channel to be the first dimension but it's the third dimension in the PIL image and Numpy array. You can reorder dimensions using [`ndarray.transpose`](https://docs.scipy.org/doc/numpy-1.13.0/reference/generated/numpy.ndarray.transpose.html). The color channel needs to be first and retain the order of the other two dimensions.

````python
def process_image(image):
    ''' Scales, crops, and normalizes a PIL image for a PyTorch model,
        returns an Numpy array
    
    Arguments:
    image -- object, PIL image to be processed
    '''
    
    # TODO: Process a PIL image for use in a PyTorch model
    
    # Resize the image
    image = image.resize((256, 256))
    
    # Crop out the center 224x224 portion of the image(left, upper, right, lower), with (0, 0) at left-upper corner
    image = image.crop((16, 16, 240, 240))
    
    # Covert PIL image to Numpy array
    image_np = np.array(image).astype(float) / 255
    
    # Normallization
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image_np = (image_np - mean) / std
    
    # Reorder the dimension to [C, H, W]
    image_np = image_np.transpose(2, 0, 1)
    
    # Covert Numpy image to tensor
    #image_tensor = torch.tensor(image_np).type(torch.FloatTensor).resize_(1,3,224,224)
    
    return image_np
````

To check your work, the function below converts a PyTorch tensor and displays it in the notebook. If your `process_image` function works, running the output through this function should return the original image (except for the cropped out portions).

````python
def imshow(image, ax=None, title=None):
    if ax is None:
        fig, ax = plt.subplots()
    
    # PyTorch tensors assume the color channel is the first dimension
    # but matplotlib assumes is the third dimension
    image = image.transpose((1, 2, 0))
    
    # Undo preprocessing
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    image = std * image + mean
    
    # Image needs to be clipped between 0 and 1 or it looks like noise when displayed
    image = np.clip(image, 0, 1)
    
    ax.imshow(image)
    
    return ax
````

## Class Prediction

Once you can get images in the correct format, it's time to write a function for making predictions with your model. A common practice is to predict the top 5 or so (usually called top-$K$) most probable classes. You'll want to calculate the class probabilities then find the $K$ largest values.

To get the top $K$ largest values in a tensor use [`x.topk(k)`](http://pytorch.org/docs/master/torch.html#torch.topk). This method returns both the highest `k` probabilities and the indices of those probabilities corresponding to the classes. You need to convert from these indices to the actual class labels using `class_to_idx` which hopefully you added to the model or from an `ImageFolder` you used to load the data ([see here](#Save-the-checkpoint)). Make sure to invert the dictionary so you get a mapping from index to class as well.

Again, this method should take a path to an image and a model checkpoint, then return the probabilities and classes.

```python
probs, classes = predict(image_path, model)
print(probs)
print(classes)
> [ 0.01558163  0.01541934  0.01452626  0.01443549  0.01407339]
> ['70', '3', '45', '62', '55']
```

````python
def predict(image_path, model, topk=5, device='cpu'):
    ''' Predict the class (or classes) of an image using a trained deep learning model.er
    
    Argument:
    image_path -- string, path of image file
    model -- object, model used to make prediction
    topk -- integer, number of result to be ranked
    device -- string, on which device to run this function
    
    Return:
    probabilities and classes -- list
    '''
    
    # TODO: Implement the code to predict the class from an image file
    
    # Turn off dropout
    model.eval()
    
    # Read PIL image file
    img = Image.open(image_path)
    
    # Process PIL image for PyTorch and convert it to a FlostTensor, move to device, resize
    image = torch.tensor(process_image(img)).to(device)
    if torch.cuda.is_available():
        image = image.type(torch.cuda.FloatTensor).resize_(1,3,224,224)
    else:
        image = image.type(torch.FloatTensor).resize_(1,3,224,224)
    
    # Forward pass of prediction
    with torch.no_grad():
        output = model.forward(image)
    
    # Model's output is log-softmax, take exponential to get the probabilities
    probs, indices = torch.exp(output).topk(topk)
    #print(indices.tolist())
    
    # Invert class_to_idx dictionary to idx_to_class
    idx_to_class = dict([(str(idx), int(cls)) for cls, idx in model.class_to_idx.items()])
    
    # Convert indices to classes with idx_to_class
    classes = [idx_to_class[str(i)] for i in indices.tolist()[0]]
    
    return probs.tolist()[0], classes
````
