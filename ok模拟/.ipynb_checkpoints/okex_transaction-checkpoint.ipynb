# Draw Picture - 正式盘

```python
# X轴
index = pd.date_range('7/15/2020', end=time.strftime('%m/%d/%Y', time.localtime(time.time())))
index = [time.strftime('%m/%d', time.localtime(e.value / 1000000000)) for e in index]
# Y轴
principal= [4872, 5057, 5271, 5347, 5484, 5627, 4961, 4820, 5290, 5675, 5167, 2213, 2213, 2213, 2213, 
           2320, 2176, 2076, 2329, 2210]
profit = [-132, 173, 211, 76, 132, 175, -670, -146, 467, 385, -508, -3166, 0, 0, 0, 105,
         -146, -98, 250, -119]
if len(index) > len(profit):
    index.pop(-1)
```

```python
plt.figure(figsize=(15, 12))
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False
plt.plot(index, principal, color='red', linewidth=1.0, marker='*', label='本金($4995)')
for a, b in zip(index, principal):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
    
plt.plot(index, profit, color='green', linewidth=1.5, marker='.', label='盈利')
for a, b in zip(index, profit):
    plt.text(a, b, b, ha='center', va='bottom', fontsize=12)
plt.legend(loc='upper left')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(u'date', fontsize=12)
plt.ylabel(u'money($)', fontsize=12)
plt.title('profit-date', fontsize=16)
```
